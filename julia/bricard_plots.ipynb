```python
import numpy as np
import tikzplotlib
import matplotlib.pyplot as plt
from book_palette import pal
```

```python
Y = np.loadtxt("bricard_ce.txt")
B = np.loadtxt("bricard_benchmark.txt")
```

```python
# Create a figure with subplots
plt.style.use('ggplot')
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(5, 5))

axs[0].plot(Y[:, 0], Y[:, 1], color=pal["lightred"])
axs[0].plot(B[:, 0], B[:, 1], color=pal["darkred"],
            linestyle='--', dashes=(2, 2), lw=2)
axs[0].set_ylabel(r'$x_2$ [m]')
axs[1].plot(Y[:, 0], Y[:, 2], color=pal["lightgreen"])
axs[1].plot(B[:, 0], B[:, 2], color=pal["darkgreen"],
            linestyle='--', dashes=(2, 2), lw=2)
axs[1].set_ylabel(r'$y_2$ [m]')
axs[2].plot(Y[:, 0], Y[:, 3], color=pal["lightblue"])
axs[2].plot(B[:, 0], B[:, 3], color=pal["darkblue"],
            linestyle='--', dashes=(2, 2), lw=2)
axs[2].set_ylabel(r'$z_2$ [m]')
axs[2].set_xlabel(r'$t$ [s]')

# Adjust the layout and spacing
fig.tight_layout()

plt.savefig('bricard_ce_1.png', dpi=450) # tikzplotlib.save("bricard_th.pgf")

# Show the plot
plt.show()
```
